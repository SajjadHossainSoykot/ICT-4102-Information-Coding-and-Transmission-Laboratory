## **Objective:**
To write a program to implement the **Shannon-Fano coding algorithm**, which is used for lossless data compression.

## **Theory:**
The **Shannon-Fano coding algorithm** is a method of constructing a binary prefix code based on a set of symbols and their probabilities. The algorithm divides the set of symbols into two parts such that the total probability of the symbols in each part is as equal as possible. It then assigns binary digits (0 and 1) to these two parts, and the process is recursively applied to each part.

Steps involved in the Shannon-Fano coding algorithm:
1. **Sort the symbols by probability** (from highest to lowest probability).
2. **Divide the list into two parts** such that the sum of the probabilities in each part is as equal as possible.
3. **Assign binary digits** to the two parts (0 to the first part, 1 to the second part).
4. **Repeat the process recursively** for each part until all symbols have a code.

The result is a binary code for each symbol, with shorter codes assigned to more probable symbols, which leads to efficient compression.

## **Algorithm:**
1. **Input the symbols and their probabilities**.
2. **Sort the symbols** in descending order of their probabilities.
3. **Divide the set into two parts** such that the sum of the probabilities in each part is as equal as possible.
4. **Assign binary digits** to the two parts (0 for the first part and 1 for the second part).
5. **Recursively apply the process** to the two parts until all symbols have codes.
6. **Output**: The binary code for each symbol.

## **Expected Output:**
1. The program will display the Shannon-Fano code for each symbol based on the input probabilities.
   
   Example input/output:
   - Shannon-Fano Codes:
   - Symbol: A, Code: 00
   - Symbol: B, Code: 01
   - Symbol: C, Code: 10
   - Symbol: D, Code: 110
   - Symbol: E, Code: 111


## **Conclusion:**
- The **Shannon-Fano coding algorithm** efficiently compresses data by assigning shorter codes to more probable symbols and longer codes to less probable symbols.
- The algorithm works by recursively dividing the set of symbols and assigning binary digits (0 or 1) to each division until all symbols have unique binary codes.
- This is a **lossless compression algorithm**, meaning the original data can be recovered from the compressed data without any loss.

## **Code:**

```python
def shannon_fano_coding(data):
    items = list(data.items())

    # Base case: only one symbol
    if len(items) == 1:
        return {items[0][0]: ""}

    # Sort by probability in descending order
    items.sort(key=lambda x: x[1], reverse=True)

    total = sum(prob for symbol, prob in items)
    cumulative = 0
    split_index = 0

    # Find split point
    for i in range(len(items)):
        cumulative += items[i][1]
        if cumulative >= total / 2:
            split_index = i
            break

    # Divide into two groups
    left_data = dict(items[:split_index + 1])
    right_data = dict(items[split_index + 1:])

    # Recursive code assignment
    codes = {}

    left_codes = shannon_fano_coding(left_data)
    for symbol in left_codes:
        codes[symbol] = "0" + left_codes[symbol]

    if right_data:
        right_codes = shannon_fano_coding(right_data)
        for symbol in right_codes:
            codes[symbol] = "1" + right_codes[symbol]

    return codes


# Input data
data = {
    'A': 0.4,
    'B': 0.3,
    'C': 0.2,
    'D': 0.05,
    'E': 0.05
}

# Generate codes
codes = shannon_fano_coding(data)

# Display output
print("Shannon-Fano Codes:")
print("Symbol\tProbability\tCode")

for symbol in data:
    print(f"{symbol}\t{data[symbol]}\t\t{codes[symbol]}")
```
