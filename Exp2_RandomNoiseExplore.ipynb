# **Experiment 2:**
To develop a program to analyze the characteristics of random noise under different mean and variance values.

## **Objective:**
To write a program to explore the characteristics of **random noise** with different **mean** and **variance**, and understand how they affect the signal quality.

## **Theory:**
In communication systems, **random noise** is typically modeled as a signal that has unpredictable values. Random noise can be generated with different characteristics, most notably its **mean** (average value) and **variance** (spread or dispersion of the signal).

1. **Mean of Noise**: The mean represents the average value of the noise. A mean of 0 corresponds to a noise signal that has no bias or offset, while a non-zero mean will cause the noise to have a constant shift in one direction.
2. **Variance of Noise**: The variance determines how spread out the noise is. A larger variance indicates more fluctuations in the noise signal, while a smaller variance corresponds to noise with smaller deviations from the mean.

By exploring random noise with different means and variances, we can analyze how these characteristics impact the signal and the **Signal-to-Noise Ratio (SNR)**.

## **Algorithm:**
1. **Generate a clean signal**: A simple sine wave is chosen as the clean signal.
2. **Add random noise with different means**: Introduce noise with varying means (0, 1, -1) to observe the impact of the noise offset.
3. **Add random noise with different variances**: Introduce noise with varying variances (0.1, 1, 10) to see how the spread of the noise affects the signal.
4. **Analyze the noisy signal**: Visualize the clean and noisy signals to observe the effects of mean and variance on the signal quality.
5. **Compute Signal-to-Noise Ratio (SNR)**: Calculate the SNR for each noise variation (mean and variance) to quantify the impact of noise.

## **Expected Output:**
1. **Graphical Output**: 
   - A series of plots showing the noisy signal with different **means** (0, 1, -1) and **variances** (0.1, 1, 10). Each plot will show how the noise affects the clean sine wave signal.
   
2. **SNR Output**:
   - Printed values for **Signal-to-Noise Ratio (SNR)** for each combination of mean and variance, quantifying the effect of different noise characteristics.
   
   Example output:

## **Code:**

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate a clean signal (a simple sine wave)
fs = 1000  # Sampling frequency
t = np.linspace(0, 1, fs)  # Time vector
freq = 5  # Frequency of the sine wave
clean_signal = np.sin(2 * np.pi * freq * t)

# Noise Characteristics
means = [0, 1, -1]  # Different means of noise
variances = [0.1, 1, 10]  # Different variances of noise

#Plot Clean Signal
plt.figure(figsize=(15,3))
plt.plot(t, clean_signal)
plt.title("Clean Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid()
plt.show()

#Plot Noise Signals in grid
fig, axes = plt.subplots(len(means),len(variances),figsize=(15,10))
snr_values={}

for i, mean in enumerate(means):
    for j, variance in enumerate (variances):
        noise = np.random.normal(mean, np.sqrt(variance),fs)
        noisy_signal = clean_signal + noise
        
        #Plot        
        axes[i,j].plot(t,noisy_signal)
        axes[i,j].set_title(f"Mean={mean}, Var={variance}")
        axes[i,j].grid()
        
        #SNR
        signal_power = np.var(clean_signal)
        noise_power = np.var(noise)
        snr=10*np.log10(signal_power/noise_power)
        snr_values[(mean,variance)] = snr
        
plt.tight_layout()
plt.show()

#Print SNR
print("\nSNR Results (dB):")
print(f"{'Mean':>5} {'Variance':>10} {'SNR (dB)':>12}")
print("-" * 30)

for (mean, variance), snr in snr_values.items():
    print(f"{mean:5} {variance:10} {snr:12.2f}")
```


## **Conclusion:**
- The **clean signal** is a smooth sine wave, unaffected by noise.
- The **mean of noise** introduces an offset to the signal, shifting it in one direction (positive or negative). A non-zero mean causes a constant drift in the signal.
- The **variance of noise** affects how spread out the noise is. A higher variance results in more fluctuations in the noise, causing greater distortion to the signal.
- The **Signal-to-Noise Ratio (SNR)** decreases as the noise variance increases, indicating more noise and poorer signal quality. Similarly, adding noise with a non-zero mean decreases the SNR, even if the variance remains constant.
- This experiment demonstrates how varying the mean and variance of random noise affects the quality of the signal and the resulting SNR.
